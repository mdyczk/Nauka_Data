```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import calmap
```

## Sprawdzanie danych

```python
df = pd.read_csv('supermarket_sales.csv') 
#Import danych z csv jako df(dataframe)
```

```python
df['Date'] = pd.to_datetime(df['Date']) 
#Zmiana typu danych kolumny Date odbject na datę
```

```python
df.dtypes
```

```python
df.set_index('Date', inplace=True) 
#Zmiana indeksu na datę
```

## Analiza

```python
sns.histplot(df['Rating'], kde=True) 
plt.axvline(x=np.mean(df['Rating']), c = 'red', ls = '--',label='srednia') 
plt.axvline(x=np.percentile(df['Rating'],25), c = 'green', ls = '--',label='25% - 75%') 
plt.axvline(x=np.percentile(df['Rating'],75), c = 'green', ls = '--') 
plt.legend()
#Histogram ocen klientów z linią trendu
```

```python
df.hist(figsize=(10,10)) 
#Histogramy wszystkich kolumn
```

```python
sns.countplot(x=df['Branch']) 
#Ilość sprzedaży w danym dziale
```

```python
sns.countplot(x=df['Payment']) 
#Rodzaje płatności
```

```python
sns.regplot(x=df['Rating'],y=df['gross income'])
#Wykres punktowy pomiędzy oceną a przychodem 
```

```python
sns.boxplot(x=df['Branch'],y=df['gross income'])
#Wykres pudełkowy pomiędzy działem a przychodem 
```

```python
df.groupby(df.index).mean(numeric_only=True) 
#Grupowanie danych przez date
```

```python
sns.lineplot(x=df.groupby(df.index).mean(numeric_only=True).index,
             y=df.groupby(df.index).mean(numeric_only=True)['gross income'])
#Wykres pomiędzy datą a przychodem
```

```python
df.duplicated().sum()
#Sprawdzenie ile jest wartości które się powtarzają
```

```python
df[df.duplicated()==True]
#Wyświetlenie powtarzających się wartości
```

```python
df.drop_duplicates(inplace=True)
#Usunięcie powtarzających się wartości
df.duplicated().sum()
#Sprawdzenie czy wszystkie się usunęły
```

```python
df.isna().sum()/len(df)
#Procent danych które są puste
```

```python
sns.heatmap(df.isnull(),cbar=False)
#Heatmapa danych pustych
```

```python
df.fillna(method='ffill',inplace=True)
#Zastąpienie pustych danych danymi z poprzedniego rzędu
df.isnull().sum()
#Zobaczenie ile wartości jest pustych
```
